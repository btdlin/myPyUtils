```python
import itertools as it

def split_candidates(candidates, needed=10):
    to_return = []
    for _, grp in it.groupby(candidates, lambda c: c.sort_weight):
        grp = list(grp)
        if len(grp) > needed:
            to_score.extend(grp)
            return to_score
        else:
            to_score.extend(grp)
            
        needed -= len(grp)
    raise ValueError('too few candidates!', needed)
        
split_candidates(candidates, 10)
```
